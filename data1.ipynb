# End-to-End Data Analysis and Machine Learning Project using Python

## Introduction

This project focuses on performing data analysis and building a simple machine learning model using Python.
It includes data cleaning, exploratory data analysis (EDA), visualization, and prediction.


## Import Libraries

The required Python libraries are imported for data processing and visualization.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

The dataset is loaded using seaborn library.


```python
df = sns.load_dataset('tips')
df.head()
```

## Data Cleaning

In this step, we check for missing values and duplicate records in the dataset.


```python
df.isnull().sum()
df.duplicated().sum()
```

## Exploratory Data Analysis (EDA)

Statistical summary is used to understand the distribution of data.


```python
df.describe()
```

## Data Visualization

Graphs are used to identify relationships between variables.


```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.show()

sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

## Correlation Analysis

Heatmap shows correlation between numerical variables.


```python
df_numeric = df.select_dtypes(include=['number'])
sns.heatmap(df_numeric.corr(), annot=True)
plt.show()
```

## Machine Learning Model

A Linear Regression model is used to predict the tip amount.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df_numeric = pd.get_dummies(df, drop_first=True)

X = df_numeric.drop('tip', axis=1)
y = df_numeric['tip']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

## Conclusion

The analysis shows that total bill has a strong impact on tip amount.
Dinner time results in higher spending compared to lunch.
The machine learning model predicts tip values with reasonable accuracy.
This project demonstrates the complete data analysis workflow.
